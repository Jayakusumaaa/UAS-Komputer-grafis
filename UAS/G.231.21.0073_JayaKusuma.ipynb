```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Data
purchase_price = [90.00, 32.00, 25.00, 50.00, 18.00, 225.00, 22.00, 18.00, 65.00, 800.00, 95.00, 17.00, 45.00]
current_price = [144.13, 65.57, 82.57, 44.56, 17.69, 833.36, 36.07, 33.24, 56.48, 897.64, 86.26, 141.64, 14.61]
gain_loss = [14269.14, 13048.57, 12302.93, 11094.20, 8824.82, 82502.64, 7176.94, 7445.54, 10391.40, 43984.36, 8539.74, 141498.36, 7287.94]

# Membuat subplot
fig, axes = plt.subplots(3, 1, figsize=(8, 10))

# Plot histogram untuk Harga Pembelian
sns.histplot(purchase_price, bins=10, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Purchase Price')

# Plot histogram untuk Harga Saat Ini
sns.histplot(current_price, bins=10, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Current Price')

# Plot histogram untuk Gain/Loss
sns.histplot(gain_loss, bins=10, kde=True, ax=axes[2])
axes[2].set_title('Distribution of Gain/Loss')

# Sesuaikan jarak antar subplot
fig.tight_layout()

# Menunjukkan plot
plt.show()
```
